#### 1.1.3 任务三：每1000行为一个数据模块，逐块读取

```python
#写入代码
chunker = pd.read_csv('train.csv', chunksize = 1000)
print(type(chunker))
for piece in chunker:
    print(type(piece))
    print(len(piece))
    print(piece.head())
```

【思考】什么是逐块读取？为什么要逐块读取呢？
【总结】对于数据量较大的文件，可考虑进行逐块读取处理。在pandas中可通过chunksize设置块的大小，可通过对其遍历进行操作；对于大文件，可以使用nrows参数读取部分
