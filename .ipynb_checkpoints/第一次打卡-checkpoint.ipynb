```python
midage = midage.reset_index(drop = True)
midage.loc[[100],['Pclass','Sex']]
```

【思考】这个reset_index()函数的作用是什么？如果不用这个函数，下面的任务会出现什么情况？

<div class="alert alert-block alert-success">
【回答】不使用reset_index()函数的结果为第一结果，对比发现，是索引的区别，该函数的具体解释如下:rop=True 就是把原来的索引index列去掉，重置index；drop=False 就是保留原来的索引，添加重置的index。两者的区别就是有没有把原来的index去掉。
</div>


#### 1.5.4 任务四：将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来

#### 1.5.5 任务五：使用iloc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来

```python
midage.iloc[[100,105,108],[2,3,4]] #无法用列名索引吗
```

<div class="alert alert-block alert-success">
【总结】首先区分下loc()和iloc()方法的区别，前者是根据索引名称确定数据，后者是根据数据位置确定数据。此时loc和iloc输出的结果相同，原因是数据框所用的索引“乘客ID”是从1开始计数的。若索引值是从0开始计数，那么iloc()方法的索引名称为100时，实际定位到101行数据。
</div>
